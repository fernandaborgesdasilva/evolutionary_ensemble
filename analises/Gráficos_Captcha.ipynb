```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("comp_time_df_results_analysis_dados_captcha.csv")
```

```python
df["file_name"].unique()
```

## Sequential

```python
# random brute-force search
bfec_rand_df = df[df["file_name"].str.contains("dados_captcha/bfec_rand_results_iter_")]
bfec_rand_df["file"] = "bfec_rand"

boxplot_bfec_rand = bfec_rand_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())

for i in range(0,10):
    bfec_rand_df_aux = bfec_rand_df[bfec_rand_df["file_name"].str.contains("dados_captcha/bfec_rand_results_iter_"+str(i))]
    last_accuracy = bfec_rand_df_aux[bfec_rand_df_aux["time_sec"]==max(bfec_rand_df_aux["time_sec"])]["ensemble_accuracy"].item()
    for t in range(max(bfec_rand_df_aux["time_sec"])+1,max(bfec_rand_df["time_sec"])+1):
        bfec_rand_df = bfec_rand_df.append({'file_name': bfec_rand_df_aux["file_name"].head(1).item(),
                                            'time_sec': t,
                                            'ensemble_accuracy': last_accuracy,
                                            'file': "bfec_rand"}, ignore_index=True)

bfec_rand_df = bfec_rand_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
bfec_rand_df.columns = ["_".join(x) for x in bfec_rand_df.columns.ravel()]
bfec_rand_df = bfec_rand_df.reset_index(level="file").reset_index(level="time_sec")

# sequential brute-force search
bfec_seq_df = df[df["file_name"].str.contains("dados_captcha/bfec_seq_results_iter_")]
bfec_seq_df["file"] = "bfec_seq"

boxplot_bfec_seq = bfec_seq_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())

for i in range(0,10):
    bfec_seq_df_aux = bfec_seq_df[bfec_seq_df["file_name"].str.contains("dados_captcha/bfec_seq_results_iter_"+str(i))]
    last_accuracy = bfec_seq_df_aux[bfec_seq_df_aux["time_sec"]==max(bfec_seq_df_aux["time_sec"])]["ensemble_accuracy"].item()
    for t in range(max(bfec_seq_df_aux["time_sec"])+1,max(bfec_seq_df["time_sec"])+1):
        bfec_seq_df = bfec_seq_df.append({'file_name': bfec_seq_df_aux["file_name"].head(1).item(),
                                          'time_sec': t,
                                          'ensemble_accuracy': last_accuracy,
                                          'file': "bfec_seq"}, ignore_index=True)

bfec_seq_df = bfec_seq_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
bfec_seq_df.columns = ["_".join(x) for x in bfec_seq_df.columns.ravel()]
bfec_seq_df = bfec_seq_df.reset_index(level="file").reset_index(level="time_sec")

# diversity
diversity_df = df[df["file_name"].str.contains("dados_captcha/diversity_results_iter_")]
diversity_df["file"] = "diversity_df"

boxplot_diversity = diversity_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())

for i in range(0,10):
    diversity_df_aux = diversity_df[diversity_df["file_name"].str.contains("dados_captcha/diversity_results_iter_"+str(i))]
    last_accuracy = diversity_df_aux[diversity_df_aux["time_sec"]==max(diversity_df_aux["time_sec"])]["ensemble_accuracy"].item()
    for t in range(max(diversity_df_aux["time_sec"])+1,max(diversity_df["time_sec"])+1):
        diversity_df = diversity_df.append({'file_name': diversity_df_aux["file_name"].head(1).item(),
                                            'time_sec': t,
                                            'ensemble_accuracy': last_accuracy,
                                            'file': "diversity_df"}, ignore_index=True)

diversity_df = diversity_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
diversity_df.columns = ["_".join(x) for x in diversity_df.columns.ravel()]
diversity_df = diversity_df.reset_index(level="file").reset_index(level="time_sec")

ax = plt.gca()
diversity_df.plot(kind='line',
                  x='time_sec',
                  y='ensemble_accuracy_mean',
                  color='#26589B',
                  label='diversity - mean accuracy',
                  ax=ax)
bfec_rand_df.plot(kind='line',
                  x='time_sec',
                  y='ensemble_accuracy_mean',
                  color='#338DFF',
                  label='random brute-force search - mean accuracy',
                  ax=ax)
bfec_seq_df.plot(kind='line',
                 x='time_sec',
                 y='ensemble_accuracy_mean',
                 color='#4F33FF',
                 label='sequential brute-force search - mean accuracy',
                 ax=ax)
ax.fill_between(diversity_df["time_sec"],
                diversity_df["ensemble_accuracy_mean"] - diversity_df["ensemble_accuracy_std"], 
                diversity_df["ensemble_accuracy_mean"] + diversity_df["ensemble_accuracy_std"],
                color='#26589B',
                alpha=.3)
ax.fill_between(bfec_rand_df["time_sec"],
                bfec_rand_df["ensemble_accuracy_mean"] - bfec_rand_df["ensemble_accuracy_std"], 
                bfec_rand_df["ensemble_accuracy_mean"] + bfec_rand_df["ensemble_accuracy_std"],
                color='#338DFF',
                alpha=.3)
ax.fill_between(bfec_seq_df["time_sec"],
                bfec_seq_df["ensemble_accuracy_mean"] - bfec_seq_df["ensemble_accuracy_std"], 
                bfec_seq_df["ensemble_accuracy_mean"] + bfec_seq_df["ensemble_accuracy_std"],
                color='#4F33FF',
                alpha=.3)
plt.rcParams["figure.figsize"] = [15,8]
plt.savefig('dados_captcha_graph_without_parallelism.png')
plt.close()
```

```python
boxplot_bfec_rand = pd.DataFrame(boxplot_bfec_rand)
boxplot_bfec_rand["exec_type"] = "random brute-force search - total execution time"

boxplot_bfec_seq = pd.DataFrame(boxplot_bfec_seq)
boxplot_bfec_seq["exec_type"] = "sequential brute-force search - total execution time"

boxplot_diversity = pd.DataFrame(boxplot_diversity)
boxplot_diversity["exec_type"] = "diversity - total execution time"

boxplot_df = pd.concat([boxplot_bfec_rand,boxplot_bfec_seq]).reset_index(level="file_name").drop(["file_name"], axis=1)
sns.boxplot(x="exec_type", y="time_sec", data=boxplot_df, palette="Set2")
plt.savefig('dados_captcha_boxplot_brut_force_results.png')
plt.close()

sns.boxplot(x="exec_type", y="time_sec", data=boxplot_diversity, palette="Set2")
plt.savefig("dados_captcha_boxplot_diversity_results.png")
plt.close()

boxplot_df_3 = pd.concat([boxplot_bfec_rand,boxplot_bfec_seq,boxplot_diversity]).reset_index(level="file_name").drop(["file_name"], axis=1)
sns.boxplot(x="exec_type", y="time_sec", data=boxplot_df_3, palette="Set2")
plt.savefig("dados_captcha_boxplot_results.png")
plt.close()
```

## Parallel

```python
def graph_parallel_gener(num_cores):
    #random brute-force search - mean accuracy
    pbfec_rand_f_name = "dados_captcha/pbfec_rand_results_iter_\d_" + str(num_cores)
    pbfec_rand_df = df[df["file_name"].str.contains(pat = pbfec_rand_f_name, regex = True)]
    pbfec_rand_df["file"] = "pbfec_rand"
    boxplot_pbfec_rand = pbfec_rand_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())

    for i in range(0,10):
        pbfec_rand_df_aux = pbfec_rand_df[pbfec_rand_df["file_name"].str.contains("dados_captcha/pbfec_rand_results_iter_"+str(i))]
        last_accuracy = pbfec_rand_df_aux[pbfec_rand_df_aux["time_sec"]==max(pbfec_rand_df_aux["time_sec"])]["ensemble_accuracy"].item()
        for t in range(max(pbfec_rand_df_aux["time_sec"])+1,max(pbfec_rand_df["time_sec"])+1):
            pbfec_rand_df = pbfec_rand_df.append({'file_name': pbfec_rand_df_aux["file_name"].head(1).item(),
                                                  'time_sec': t,
                                                  'ensemble_accuracy': last_accuracy,
                                                  'file': "pbfec_rand"}, ignore_index=True)
    pbfec_rand_df = pbfec_rand_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
    pbfec_rand_df.columns = ["_".join(x) for x in pbfec_rand_df.columns.ravel()]
    pbfec_rand_df = pbfec_rand_df.reset_index(level="file").reset_index(level="time_sec")

    #sequential brute-force search - mean accuracy
    pbfec_seq_f_name = "dados_captcha/pbfec_seq_results_iter_\d_" + str(num_cores)
    pbfec_seq_df = df[df["file_name"].str.contains(pat = pbfec_seq_f_name, regex = True)]
    pbfec_seq_df["file"] = "pbfec_seq"
    boxplot_pbfec_seq = pbfec_seq_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())

    for i in range(0,10):
        pbfec_seq_df_aux = pbfec_seq_df[pbfec_seq_df["file_name"].str.contains("dados_captcha/pbfec_seq_results_iter_"+str(i))]
        last_accuracy = pbfec_seq_df_aux[pbfec_seq_df_aux["time_sec"]==max(pbfec_seq_df_aux["time_sec"])]["ensemble_accuracy"].item()
        for t in range(max(pbfec_seq_df_aux["time_sec"])+1,max(pbfec_seq_df["time_sec"])+1):
            pbfec_seq_df = pbfec_seq_df.append({'file_name': pbfec_seq_df_aux["file_name"].head(1).item(),
                                                'time_sec': t,
                                                'ensemble_accuracy': last_accuracy,
                                                'file': "pbfec_seq"}, ignore_index=True)
    pbfec_seq_df = pbfec_seq_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
    pbfec_seq_df.columns = ["_".join(x) for x in pbfec_seq_df.columns.ravel()]
    pbfec_seq_df = pbfec_seq_df.reset_index(level="file").reset_index(level="time_sec")
    
    #diversity - mean accuracy
    parallel_diversity_f_name = "dados_captcha/parallel_diversity_results_iter_\d_" + str(num_cores)
    parallel_diversity_df = df[df["file_name"].str.contains(pat=parallel_diversity_f_name, regex=True)]
    parallel_diversity_df["file"] = "parallel_diversity"
    boxplot_parallel_diversity = parallel_diversity_df.groupby('file_name')['time_sec'].apply(lambda g: g.max() - g.min())
    
    for i in range(0,10):
        parallel_diversity_df_aux = parallel_diversity_df[parallel_diversity_df["file_name"].str.contains("dados_captcha/parallel_diversity_results_iter_"+str(i))]
        last_accuracy = parallel_diversity_df_aux[parallel_diversity_df_aux["time_sec"]==max(parallel_diversity_df_aux["time_sec"])]["ensemble_accuracy"].item()
        for t in range(max(parallel_diversity_df_aux["time_sec"])+1,max(parallel_diversity_df["time_sec"])+1):
            parallel_diversity_df = parallel_diversity_df.append({'file_name': parallel_diversity_df_aux["file_name"].head(1).item(),
                                                                'time_sec': t,
                                                                'ensemble_accuracy': last_accuracy,
                                                                'file': "parallel_diversity"}, ignore_index=True)
    parallel_diversity_df = parallel_diversity_df.drop(["file_name"], axis=1).groupby(["file","time_sec"]).agg(["mean","std"])
    parallel_diversity_df.columns = ["_".join(x) for x in parallel_diversity_df.columns.ravel()]
    parallel_diversity_df = parallel_diversity_df.reset_index(level="file").reset_index(level="time_sec")


    ax = plt.gca()

    parallel_diversity_df.plot(kind='line',
                               x='time_sec',
                               y='ensemble_accuracy_mean',
                               color='#26589B',
                               label='diversity - mean accuracy',
                               ax=ax)
    pbfec_rand_df.plot(kind='line',
                       x='time_sec',
                       y='ensemble_accuracy_mean',
                       color='#338DFF',
                       label='random brute-force search - mean accuracy',
                       ax=ax)
    pbfec_seq_df.plot(kind='line',
                      x='time_sec',
                      y='ensemble_accuracy_mean',
                      color='#4F33FF',
                      label='sequential brute-force search - mean accuracy',
                      ax=ax)
    ax.fill_between(parallel_diversity_df["time_sec"],
                    parallel_diversity_df["ensemble_accuracy_mean"] - parallel_diversity_df["ensemble_accuracy_std"], 
                    parallel_diversity_df["ensemble_accuracy_mean"] + parallel_diversity_df["ensemble_accuracy_std"],
                    color='#26589B',
                    alpha=.3)
    ax.fill_between(pbfec_rand_df["time_sec"],
                    pbfec_rand_df["ensemble_accuracy_mean"] - pbfec_rand_df["ensemble_accuracy_std"], 
                    pbfec_rand_df["ensemble_accuracy_mean"] + pbfec_rand_df["ensemble_accuracy_std"],
                    color='#338DFF',
                    alpha=.3)
    ax.fill_between(pbfec_seq_df["time_sec"],
                    pbfec_seq_df["ensemble_accuracy_mean"] - pbfec_seq_df["ensemble_accuracy_std"], 
                    pbfec_seq_df["ensemble_accuracy_mean"] + pbfec_seq_df["ensemble_accuracy_std"],
                    color='#4F33FF',
                    alpha=.3)
    plt.rcParams["figure.figsize"] = [15,8]
    save_file_name = 'dados_captcha_' + str(num_cores) + '_cores_graph_with_parall.png'
    plt.savefig(save_file_name)
    plt.close()

    boxplot_pbfec_rand = pd.DataFrame(boxplot_pbfec_rand)
    boxplot_pbfec_rand["exec_type"] = "random brute-force search - total execution time"

    boxplot_pbfec_seq = pd.DataFrame(boxplot_pbfec_seq)
    boxplot_pbfec_seq["exec_type"] = "sequential brute-force search - total execution time"

    boxplot_parallel_diversity = pd.DataFrame(boxplot_parallel_diversity)
    boxplot_parallel_diversity["exec_type"] = "diversity - total execution time"

    boxplot_df = pd.concat([boxplot_pbfec_rand,boxplot_pbfec_seq]).reset_index(level="file_name").drop(["file_name"], axis=1)
    sns.boxplot(x="exec_type", y="time_sec", data=boxplot_df, palette="Set2")
    plt.savefig('dados_captcha_' + str(num_cores) + "_cores_boxplot_brut_force_results.png")
    plt.close()

    sns.boxplot(x="exec_type", y="time_sec", data=boxplot_parallel_diversity, palette="Set2")
    plt.savefig('dados_captcha_' + str(num_cores) + "_cores_boxplot_diversity_results.png")
    plt.close()

    boxplot_df_3 = pd.concat([boxplot_pbfec_rand,boxplot_pbfec_seq,boxplot_parallel_diversity]).reset_index(level="file_name").drop(["file_name"], axis=1)
    sns.boxplot(x="exec_type", y="time_sec", data=boxplot_df_3, palette="Set2")
    plt.savefig('dados_captcha_' + str(num_cores) + "_cores_boxplot_results.png")
    plt.close()
```
