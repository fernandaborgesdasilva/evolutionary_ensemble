```python
import pandas as pd
```

```python
df = pd.read_csv("results_analysis_dados_captcha.csv")
```

```python
comp_time_df = pd.DataFrame(columns=['file_name', 'time_sec', 'ensemble_accuracy'])

for file in pd.unique(df["file_name"]):
    print(file)
    last_accuracy = 0
    aux_comp_time_df = df[df["file_name"] == file]
    for t in range(1,max(aux_comp_time_df["time_sec"])+1):
        try:
            last_accuracy = aux_comp_time_df[aux_comp_time_df["time_sec"]==t]["ensemble_accuracy"].item()
        except: pass
        comp_time_df = comp_time_df.append({'file_name': file,
                                            'time_sec': t,
                                            'ensemble_accuracy': last_accuracy}, ignore_index=True)
```

```python
comp_time_df.to_csv("comp_time_df_results_analysis_dados_captcha.csv", index = None, header=True)
```
